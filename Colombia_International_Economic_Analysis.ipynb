```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Inflow.csv")
df2 = pd.read_csv("wages.csv")
df3 = pd.read_csv("growth rate output per worker.csv")
df4 = pd.read_csv("real relative wages.csv")
```

```python
df.head(100)
```

```python
outflows = df[(df['Country of birth/nationality'] == 'Colombia') & (df['Variable'] == 'Outflows of foreign population by nationality')]
inflows = df[(df['Country of birth/nationality'] == 'Colombia') & (df['Variable'] == 'Inflows of foreign population by nationality')]

avg_outflows = outflows.groupby('Year')['Value'].mean().reset_index()
avg_inflows = inflows.groupby('Year')['Value'].mean().reset_index()

plt.figure(figsize=(10, 6))

plt.plot(avg_inflows['Year'], avg_inflows['Value'], marker='o', color='blue', label='Average Outflows')
plt.plot(avg_outflows['Year'], avg_outflows['Value'], marker='o', color='red', label='Average Inflows')

plt.title('Average Inflows and Outflows from Colombia per Year')
plt.xlabel('Year')
plt.ylabel('Average Population')
plt.grid(True)

plt.legend()

plt.show()
```

```python
average_value_per_country_year = df3.groupby(['ref_area', 'time'])['obs_value'].mean().reset_index()

average_value_per_country_year.head(1000)
```

```python
filtered_df = average_value_per_country_year[df3['ref_area'].isin(['COL'])]

plt.figure(figsize=(15, 8))

for country in filtered_df['ref_area'].unique():
    country_data = filtered_df[filtered_df['ref_area'] == country]
    plt.plot(country_data['time'], country_data['obs_value'], marker='o', linestyle='-', label=country)

plt.title('Change in Wage per Year in Colombia')
plt.xlabel('Year')
plt.ylabel('Wage Change Percent')
plt.grid(True)
plt.xticks(range(filtered_df['time'].min(), filtered_df['time'].max() + 1))
plt.legend(title='Country')
plt.show()
```

```python
grouped_df = df4.groupby(['Country', 'TIME'])['Value'].mean().reset_index()

grouped_df.head(500)
```

```python
colombia_only_df = grouped_df[grouped_df['Country'] == 'Colombia']

# Plotting the data for Colombia
plt.figure(figsize=(10, 5))
plt.plot(colombia_only_df['TIME'], colombia_only_df['Value'], marker='o', linestyle='-', color='b')
plt.title('Value per Year for Colombia')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
